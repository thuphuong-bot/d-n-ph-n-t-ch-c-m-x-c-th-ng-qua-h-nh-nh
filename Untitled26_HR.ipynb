3.2 Sự tương quan giữa các biến dữ liệu

```python
numerical_features = dt.select_dtypes(include=np.number)

# Calculate the correlation matrix
corr = numerical_features.corr()

# Display the correlation matrix
corr
```

```python
sns.set(style = 'white')
plt.figure(figsize=(10,8))
mask = np.zeros_like(corr, dtype=np.bool)

mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr,mask = mask,  annot = True,square=True, cmap='coolwarm', xticklabels= True, yticklabels = True, cbar_kws={'shrink': .5})
```

3.8 Số năm làm việc và số dự án có ảnh hưởng tới sự rời đi của nhân viên

```python
bins = np.linspace(1.5, 7.5 , 7)
plt.hist(dt[dt['left']==1]['number_project'], bins = bins, alpha =1, label = "Employee left" )
plt.hist(dt[dt['left']==0]['number_project'], bins = bins, alpha=0.5, label = 'Employee stayed')
plt.title('number of project')
```

```python
plt.figure(figsize =(7,5))
bins = np.linspace(1.0, 11,10)
plt.hist(dt[dt['left']==1]['time_spend_company'], bins=bins, alpha=1, label='Employees Left')
plt.hist(dt[dt['left']==0]['time_spend_company'], bins=bins, alpha = 0.5, label = 'Employee Stayed')
plt.grid(axis='x')
plt.xticks(np.arange(2,11))
plt.xlabel('time_spend_company')
plt.title('Years in the company', fontsize=14)
plt.legend(loc='best');
```

3.9 Ảnh hưởng của hiệu suất làm việc

```python
plt.figure(figsize =(7,5))
bins = np.linspace(0.3, 1.000, 14)
plt.hist(dt[dt['left']==1]['last_evaluation'], bins = bins , alpha = 1, label = 'Employees left')
plt.hist(dt[dt['left']==0]['last_evaluation'], bins = bins, alpha = 0.5, label = 'Employees stayed')
plt.grid(axis = 'x')
plt.xlabel('last_evaluation')
plt.title('Last evaluation, left')
plt.legend(loc = 'best')
```
